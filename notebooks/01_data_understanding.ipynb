## Load Raw CSV Files

```python
customers = pd.read_csv("../data/raw/customers.csv")
products = pd.read_csv("../data/raw/products.csv")
orders = pd.read_csv("../data/raw/orders.csv")
order_items = pd.read_csv("../data/raw/order_items.csv")
deliveries = pd.read_csv("../data/raw/deliveries.csv")
support_tickets = pd.read_csv("../data/raw/support_tickets.csv")
```

## Check Missing Values

```python
print("Missing values in customers:")
print(customers.isnull().sum())

print("\nMissing values in products:")
print(products.isnull().sum())

print("\nMissing values in orders:")
print(orders.isnull().sum())

print("\nMissing values in order_items:")
print(order_items.isnull().sum())

print("\nMissing values in deliveries:")
print(deliveries.isnull().sum())

print("\nMissing values in support_tickets:")
print(support_tickets.isnull().sum())
```

## Check Duplicate Rows

```python
print("Duplicate rows in customers:", customers.duplicated().sum())
print("Duplicate rows in products:", products.duplicated().sum())
print("Duplicate rows in orders:", orders.duplicated().sum())
print("Duplicate rows in order_items:", order_items.duplicated().sum())
print("Duplicate rows in deliveries:", deliveries.duplicated().sum())
print("Duplicate rows in support_tickets:", support_tickets.duplicated().sum())
```

## Initial Data Understanding Observations

2. The products dataset contains realistic product names, brands, MRP, selling price, discount percentage, cost price, seller name, product rating, and rating count.
3. Some columns have missing values, especially delivery-related and support ticket-related columns.
4. There are duplicate rows in some datasets.
5. Some text columns may need cleaning because values may contain extra spaces or inconsistent formatting.
6. Some numeric columns need validation, such as quantity, price, delay_days, product_rating, and customer_satisfaction_score.
