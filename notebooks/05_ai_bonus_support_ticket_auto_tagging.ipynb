# 05 AI Bonus: Support Ticket Auto-Tagging

This is a simple AI-style bonus feature for the e-commerce analytics project.

The goal is to automatically suggest:
- Issue category
- Ticket priority

based on the customer issue description.

This is not a production AI model.  
It is a beginner-friendly automation using keyword-based rules.

```python
import pandas as pd
import os
```

## Load Support Ticket Data

```python
support_tickets = pd.read_csv("../data/cleaned/support_tickets_cleaned.csv")

support_tickets.head()
```

## Create Simple Auto-Tagging Rules

These functions read the issue description and predict the issue category and priority.

```python
def predict_issue_category(description):
    text = str(description).lower()

    if any(word in text for word in ["refund", "money back", "amount not received"]):
        return "Refund"
    elif any(word in text for word in ["delivery", "delayed", "late", "not delivered", "courier"]):
        return "Delivery"
    elif any(word in text for word in ["return", "replace", "exchange", "wrong item", "wrong model"]):
        return "Return"
    elif any(word in text for word in ["payment", "upi", "card", "transaction", "checkout"]):
        return "Payment"
    elif any(word in text for word in ["defective", "damaged", "quality", "product", "not working"]):
        return "Product"
    elif any(word in text for word in ["app", "website", "login", "page", "error"]):
        return "App/Website"
    else:
        return "Other"
```

```python
def predict_priority(description):
    text = str(description).lower()

    if any(word in text for word in ["not delivered", "payment failed", "refund not received", "damaged", "defective"]):
        return "High"
    elif any(word in text for word in ["delayed", "late", "return", "exchange", "wrong item", "wrong model"]):
        return "Medium"
    else:
        return "Low"
```

## Apply Auto-Tagging

```python
support_tickets_ai = support_tickets.copy()

support_tickets_ai["predicted_issue_category"] = support_tickets_ai["issue_description"].apply(predict_issue_category)
support_tickets_ai["predicted_priority"] = support_tickets_ai["issue_description"].apply(predict_priority)

support_tickets_ai.head()
```

## Compare Predicted Values with Actual Values

```python
support_tickets_ai["category_match"] = (
    support_tickets_ai["issue_category"].astype(str).str.strip().str.lower()
    == support_tickets_ai["predicted_issue_category"].astype(str).str.strip().str.lower()
)

support_tickets_ai["priority_match"] = (
    support_tickets_ai["priority"].astype(str).str.strip().str.lower()
    == support_tickets_ai["predicted_priority"].astype(str).str.strip().str.lower()
)

category_match_rate = support_tickets_ai["category_match"].mean() * 100
priority_match_rate = support_tickets_ai["priority_match"].mean() * 100

print("Category Match Rate:", round(category_match_rate, 2), "%")
print("Priority Match Rate:", round(priority_match_rate, 2), "%")
```

## Check Prediction Counts

```python
support_tickets_ai["predicted_priority"].value_counts()
```

## Save Auto-Tagged Output

```python
os.makedirs("../ai_bonus", exist_ok=True)

support_tickets_ai.to_csv("../ai_bonus/ticket_auto_tagging_output.csv", index=False)

print("AI bonus output saved successfully.")
```
