# 04 Exploratory Data Analysis

This notebook is used to explore the cleaned e-commerce dataset.

EDA means Exploratory Data Analysis.  
In this step, I will check sales trends, product performance, returns, delivery delays, and customer support issues.

This notebook uses simple pandas and matplotlib code.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Cleaned Datasets

```python
customers = pd.read_csv("../data/cleaned/customers_cleaned.csv")
products = pd.read_csv("../data/cleaned/products_cleaned.csv")
orders = pd.read_csv("../data/cleaned/orders_cleaned.csv")
order_items = pd.read_csv("../data/cleaned/order_items_cleaned.csv")
deliveries = pd.read_csv("../data/cleaned/deliveries_cleaned.csv")
support_tickets = pd.read_csv("../data/cleaned/support_tickets_cleaned.csv")
```

## Convert Date Columns

```python
orders["order_date"] = pd.to_datetime(orders["order_date"])
deliveries["expected_delivery_date"] = pd.to_datetime(deliveries["expected_delivery_date"])
deliveries["actual_delivery_date"] = pd.to_datetime(deliveries["actual_delivery_date"])
support_tickets["ticket_date"] = pd.to_datetime(support_tickets["ticket_date"])
```

## Basic Business KPIs

```python
total_sales = order_items["revenue"].sum()
total_profit = order_items["profit"].sum()
total_orders = orders["order_id"].nunique()
total_customers = customers["customer_id"].nunique()
average_order_value = total_sales / total_orders

print("Total Sales:", round(total_sales, 2))
print("Total Profit:", round(total_profit, 2))
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
print("Average Order Value:", round(average_order_value, 2))
```

### Observation

The basic KPI analysis gives an overall view of total sales, profit, orders, customers, and average order value.

## Monthly Sales Trend

```python
orders["order_month"] = orders["order_date"].dt.to_period("M").astype(str)

sales_data = order_items.merge(
    orders[["order_id", "order_date", "order_month", "region", "city", "order_status"]],
    on="order_id",
    how="left"
)

monthly_sales = sales_data.groupby("order_month")["revenue"].sum().reset_index()

monthly_sales
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales["order_month"], monthly_sales["revenue"], marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

### Observation

Monthly sales trend helps identify which months had higher or lower sales.

## Category-Wise Sales

```python
sales_product_data = sales_data.merge(
    products[["product_id", "product_name", "category", "sub_category", "brand"]],
    on="product_id",
    how="left"
)

category_sales = sales_product_data.groupby("category")["revenue"].sum().sort_values(ascending=False)

category_sales
```

```python
category_sales.plot(kind="bar", figsize=(8, 5))
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

### Observation

Category-wise sales shows which product categories contribute more to total revenue.

## Region-Wise Sales

```python
region_sales = sales_product_data.groupby("region")["revenue"].sum().sort_values(ascending=False)

region_sales
```

```python
region_sales.plot(kind="bar", figsize=(8, 5))
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

### Observation

Region-wise sales helps understand which regions are performing better.

## Top 10 Products by Sales

```python
top_products = sales_product_data.groupby("product_name")["revenue"].sum().sort_values(ascending=False).head(10)

top_products
```

```python
top_products.plot(kind="barh", figsize=(10, 6))
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product Name")
plt.show()
```

### Observation

Top product analysis helps the business understand which products are generating the most sales.

## Return Analysis

```python
return_count = order_items["return_status"].value_counts()

return_count
```

```python
return_reasons = order_items[order_items["return_status"] == "Returned"]["return_reason"].value_counts()

return_reasons
```

```python
return_reasons.plot(kind="bar", figsize=(8, 5))
plt.title("Return Reasons")
plt.xlabel("Return Reason")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### Observation

Return reason analysis helps identify why customers are returning products.

## Delivery Delay Analysis

```python
delivery_status_count = deliveries["delivery_status"].value_counts()

delivery_status_count
```

```python
delay_by_partner = deliveries[
    deliveries["delivery_status"] == "Delayed"
].groupby("delivery_partner")["order_id"].count().sort_values(ascending=False)

delay_by_partner
```

```python
delay_by_partner.plot(kind="bar", figsize=(8, 5))
plt.title("Delayed Orders by Delivery Partner")
plt.xlabel("Delivery Partner")
plt.ylabel("Delayed Orders")
plt.xticks(rotation=45)
plt.show()
```

### Observation

Delivery delay analysis helps identify which delivery partners need improvement.

## Customer Support Analysis

```python
tickets_by_category = support_tickets["issue_category"].value_counts()

tickets_by_category
```

```python
tickets_by_category.plot(kind="bar", figsize=(8, 5))
plt.title("Support Tickets by Issue Category")
plt.xlabel("Issue Category")
plt.ylabel("Ticket Count")
plt.xticks(rotation=45)
plt.show()
```

```python
ticket_status_count = support_tickets["ticket_status"].value_counts()

ticket_status_count
```

```python
avg_resolution = support_tickets.groupby("issue_category")["resolution_days"].mean().sort_values(ascending=False)

avg_resolution
```

```python
avg_satisfaction = support_tickets["customer_satisfaction_score"].mean()

print("Average Customer Satisfaction Score:", round(avg_satisfaction, 2))
```

### Observation

Customer support analysis helps identify common complaint categories and support performance.
